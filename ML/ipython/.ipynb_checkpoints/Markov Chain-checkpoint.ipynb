2. 일본어, 중국어 : 모든 단어를 붙인다.
3. 한국어 : 띄어쓰기 규칙이 복잡해, 맞춤법 검사기 사용

```python
from konlpy.tag import Okt
from bs4 import BeautifulSoup
import urllib.request
import os, re, json, random

# 마르코프 체인 딕셔너리 만들기 
def make_dic(words):
    tmp = ["@"]
    dic = {}
    for word in words : 
        tmp.append(word)
        if len(tmp) < 3 : continue
        if len(tmp) > 3 : tmp = tmp[1:]
        set_word3(dic, tmp)
        
        if word == ".":
            tmp=["@"]
            continue
    
    return dic


# 딕셔너리에 데이터 등록
def set_word3(dic, s3):
    w1, w2, w3 = s3
    if not w1 in dic: dic[w1] = {}
    if not w2 in dic[w1]: dic[w1][w2] = {}
    if not w3 in dic[w1][w2]: dic[w1][w2][w3] = 0
    dic[w1][w2][w3] += 1
    

# 문장 만들기 
def make_sentence(dic):
    ret = []
    if not "@" in dic : return "no dic"
    top = dic["@"]
    w1 = word_choice(top)
    w2 = word_choice(top[w1])
    ret.append(w1)
    ret.append(w2)
    
    while True :
        w3 = word_choice(dic[w1][w2])
        ret.append(w3)
        if w3 == ".": break
        w1, w2 = w2, w3
    
    ret = "".join(ret)
    
    # 띄어쓰기
    params = urllib.parse.urlencode({
        "_callback":"",
        "q":ret
    })
    
    headers = { 'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36' }

    
    # 네이버 맞춤법 검사기 사용
    base_url = 'https://m.search.naver.com/p/csearch/ocontent/spellchecker.nhn'
    data = urllib.request.urlopen("https://m.search.naver.com/p/csearch/dcontent/spellchecker.nhn?"+params)
    data = data.read().decode("utf-8")[1:-2]
    data = json.loads(data)
    data = data["message"]["result"]["html"]
    data = soup = BeautifulSoup(data, "html.parser").getText()
    
    # 리턴
    return data 


def word_choice(sel):
    keys = sel.keys()
    return random.choice(list(keys))


# 문장 읽기 
toji_file = "../dataset/fictions/tozi_1.txt"
dict_file = "markov-toji.json"

if not os.path.exists(dict_file):
    # 토지 텍스트 파일 읽어오기 
    with open(toji_file, "r", encoding="utf-8") as f : 
        text = f.read()
        text.replace("…", "")
        
        # 형태소 분석
        okt = Okt()
        malist = okt.pos(text, norm=True)
        words = []
        for word in malist : 
            # 구두점 등은 대상에서 제외(단, 마침표는 포함)
            if not word[1] in ["Punctuation"]:
                words.append(word[0])
                
            if word[0] == "." : 
                words.append(word[0])
        
        # 딕셔너리 생성
        dic = make_dic(words)
        json.dump(dic, open(dict_file, "w", encoding="utf-8"))
else : 
    dic = json.load(open(dict_file, "r"))
    
    
# 문장 만들기
for i in range(3):
    s = make_sentence(dic)
    print(s)
    print("---")
        

```

```python
import re
import requests

URL = 'https://m.search.naver.com/p/csearch/dcontent/spellchecker.nhn?'

def spellCorrection(q): 
    result = None 
    
    params = { '_callback': 'window.__jindo2_callback._spellingCheck_0', 'q': q, } 
    headers = { 'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/57.0.2987.133 Safari/537.36' } 
    
    response = requests.get(URL, params=params, headers=headers).text 
    
    errata_count = re.findall( 
        #errata_count 파싱 정규식 
        r'"errata_count":"?(.*)"?,',
        response 
    )[0] 

    result = re.sub( 
        #span 태그 제거 정규식 
        r"<[^>]*>", 
        "", 
        re.findall( 
            #교정 결과 파싱 정규식 
            r'"html":"(.*)"', 
            response 
        )[0] 
    ) 

    return errata_count, result

#동작 테스트용 
if __name__ == '__main__': 
    q = "아버쥐가방에들어가신다" 
    ec, res = spellCorrection(q) 
    print(ec, res)

https://leichtjoon.tistory.com/45
```
