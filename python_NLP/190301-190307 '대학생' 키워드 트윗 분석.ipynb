```python
import nltk 
from konlpy.tag import Okt; t = Okt()
```

```python
tw_text = ''

for each_line in tw :
    tw_text = tw_text + each_line
```

```python
tokens_ko = t.morphs(tw_text)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name='tweet')
```

```python
ko = nltk.Text(tokens_ko, name='tweet')
ko.vocab().most_common(100)
```

```python
stop_words = ['이', '의', '을', '에', '들', '가', '인데', '라고', 'ㅜ', '엔', 'ㅊㅐ', '저', '못',
             '내', '기', '대', '하고', '만', '네', '와', '적', '시', '이랑', '아니', '뭐', '하고', '분', '해'
             ,'수', '거', '와', '야', '가', '27', '에서', 'x', '적', '.......', '하고', '하면',
              '&#', '(', ')', ':', '는', '은', '으로', '한다', '한', '에서', '3', '을', '할', '/', '두',
             '를', '고','로','도','과','지','인','님','잘','다','어','아', '이라', 
             '그', '나', '너무', '진짜', '보고', '지금', '해서', 'ㅋㅋ', '원', '알', '장',
             '안', '팅', '때', '공', '이런', '대학생', '입니다']

tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]

ko = nltk.Text(tokens_ko, name='tweet')
ko.vocab().most_common(30)
```

```python
from matplotlib import font_manager, rc 
import pandas as pd
import numpy as np
import platform
import matplotlib.pyplot as plt 

%matplotlib inline

path = "C:/Windows/Fonts/malgun.ttf"

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else :
    print('Unknown system. Sooooorry.')

plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(15, 6))
ko.plot(50)
plt.show()
```
