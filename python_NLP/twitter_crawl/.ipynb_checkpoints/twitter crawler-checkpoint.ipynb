## Import Libraries

```python
import requests, re, time, random
import datetime as dt
import pandas as pd 
from bs4 import BeautifulSoup as bs 
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome('C:/chrome/chromedriver.exe')
```

## Setting Constant

```python
KEYWORD = 'KIA'
START_DATE = dt.date(year=2019, month=2, day=15)
UNTIL_DATE = dt.date(year=2019, month=2, day=16)
END_DATE = dt.date(year=2019, month=3, day=4)
SLEEP_RANDOM = random.randint(1,3)
GET_HEIGHT = "return document.body.scrollHeight"
MOVE_SCROLL = "window.scrollTo(0, document.body.scrollHeight);"
```

```python
def parse() :
    html = driver.page_source
    soup = bs(html, 'html.parser')  
    
    return soup

def control_script(statement) :

    pattern = re.compile(r'return')
    is_matched = re.match(pattern, statement)
    
    if is_matched :
        result = driver.execute_script(statement)
        return result

    else : 
        driver.execute_script(statement)
```

## Crawling Code

```python
frequencies = [] 

while END_DATE != START_DATE :
    URL = f'https://twitter.com/search?q={KEYWORD}%20since%3A{str(START_DATE)}%20until%3A{str(UNTIL_DATE)}&src=typd'
    driver.get(URL)
    time.sleep(SLEEP_RANDOM)
    soup = parse()
    
    currHeight = control_script(GET_HEIGHT)
    
    while True : 
        
        daily_freq = {'Date':START_DATE}
        word_freq = 0
        tweets = soup.find_all("p", {"class" : "TweetTextSize"})
        word_freq += len(tweets)
        
        control_script(MOVE_SCROLL)
        time.sleep(SLEEP_RANDOM)
        
        lastHeight = control_script(GET_HEIGHT)
        
        if lastHeight != currHeight : #현재 위치가 스크롤의 맨 마지막인가? 
            
            soup = parse()
            tweets = soup.find_all("p", {"class" : "TweetTextSize"})
            word_freq = len(tweets)
            
        else : 
            daily_freq['Frequency']=word_freq
            frequencies.append(daily_freq)
            
            #initialize 
            word_freq = 0 
            START_DATE = UNTIL_DATE 
            daily_freq={} 

            UNTIL_DATE += dt.timedelta(days=1)
            break #else break
        
        #else문 실행이 안되면 마지막 height를 현재 height에 대입 
        currHeight = lastHeight
```

```python
df = pd.DataFrame(frequencies)
df.head()

import matplotlib.pyplot as plt 

plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
# plt.scatter(df.Date, df.Frequency)
plt.plot(df.Date, df.Frequency)
```
