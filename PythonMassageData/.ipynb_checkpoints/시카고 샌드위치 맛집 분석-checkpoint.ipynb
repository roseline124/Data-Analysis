# <font color="#ff8104">맛집 위치 지도에 표시하기</font>

```python
import folium
import pandas as pd
import googlemaps
import numpy as np
```

```python
df = pd.read_csv("file", index_col=0)
df.head()
```

```python
gmaps_key = "구글맵 키 입력"
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
lat = []
lng = []

for n in tqdm_notebook(df.index) : 
    if df['Address'][n] != 'Multiple' : 
        target_name = df['Address'][n] + ', ' + 'Cichago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        
    else :
        lat.append(np.nan)
        lng.append(np.nan)        
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()],
                     zoom_start=11)
folium.Marker([df['lat'].mean(), df['lng'].mean()],
                 popup='center').add_to(mapping)

mapping
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()],
                     zoom_start=11)

for n in df.index : 
    if df['Address'][n] != 'Multiple' :
        
        folium.Marker([df['lat'][n], df['lng'][n]],
                         popup=df['Cafe'][n]).add_to(mapping)

mapping
```

# <font color="#ff8104">네이버 영화 평점 데이터 수집</font>

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
from urllib.request import urlopen
```

```python
URL = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20190309"
html = urlopen(URL)
soup = bs(html, 'html.parser')
soup
```

```python
soup.find_all('div', 'tit5')
```

```python
soup.find_all('div', 'tit5')[0].a.string
```

```python
soup.find_all('td', 'point')[0].string
```

```python
date = pd.date_range('2019-02-01', periods=30, freq='D')
date
```

```python
import urllib
from tqdm import tqdm_notebook

movie_date = []
movie_name = []
movie_point = []

for today in tqdm_notebook(date) : 
    html = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    response = urlopen(html.format(date=
                                  urllib.parse.quote(today.strftime('%Y%m%d'))))
    
    soup = bs(response, 'html.parser')
    
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])    
```

```python
movie = pd.DataFrame( 
            {'date':movie_date, 'name' : movie_name, 'point':movie_point} )

movie.head()
```

```python
import numpy as np

movie_unique = pd.pivot_table(movie, index=['name'], aggfunc=np.sum)
movie_best = movie_unique.sort_values(by='point', ascending=False)
movie_best.head()
```

```python
def round_point(column) : 
    column = column[:6]
    return round(float(column), 2)

movie_best['point'] = movie_best['point'].apply(round_point)
movie_best.head()
```

```python
def multiple_point(column) :
    return int(column*100) 

movie_best['point'] = movie_best['point'].apply(multiple_point)
movie_best.head()
```

```python
tmp = movie.query('name==["보헤미안 랩소디"]')

#type 변환
tmp['point'] = tmp['point'].astype('float')
tmp.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,8))

plt.plot(tmp['date'], tmp['point'])
plt.legend(loc='best')
plt.grid()
plt.show()
```
